```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,15,30)
y= np.sin(x)+0.1*np.random.randn(len(x))
plt.plot(x,y)
```

```python
#to change how the line looks in the console so that there are points at each actual data point and so that the line is dashed 
plt.plot(x,y,"o--")
```

```python
#to change color 
plt.plot(x,y,color = 'purple')
```

```python
#to change line width
plt.plot(x,y,lw = 5.0)
```

```python
#to change figure size
plt.figure(figsize = (4,3))
plt.plot(x,y,lw = 5.0)
```

```python
#to add axis labels and title
plt.figure(figsize = (4,3))
plt.plot(x,y,'o--',lw = 2.0, color = 'purple')
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.title('Voltage versus Time')
```

```python
#adding more labels and a legend
plt.figure(figsize = (4,3))
plt.plot(x,y,'o--',lw = 2.0, color = 'purple', label = "Component 1")
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.title('Voltage versus Time')
plt.legend(loc='upper right', fontsize =15)
```

```python
x2 = np.linspace(0,15,100)
y2 = np.sin(x2)
```

```python
# easy to show two graphs at the same time 
plt.figure(figsize = (10,3))
plt.plot(x,y,'o',lw = 2.0, color = 'purple', label = "Component 1")
plt.plot(x2,y2,"-", color = 'green', lw = 2.0, label = 'Component 2') 
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.title('Voltage versus Time')
plt.legend(loc='upper right', fontsize =8)
```

```python
# adjusting limits of graph 
plt.figure(figsize = (10,3))
plt.plot(x,y,'o',lw = 2.0, color = 'purple', label = "Data")
plt.plot(x2,y2,"-", color = 'green', lw = 2.0, label = 'Theory') 
plt.xlabel('Time[s]')
plt.ylabel('Voltage[V]')
plt.title('Voltage versus Time')
plt.ylim(top =2.0) #this line is key here 
plt.legend(loc='upper right', fontsize =8) #fontsize also changes how big the legend is
```

```python
res_a1 = 0.2*np.random.randn(1000)+0.4
res_b1 = 0.25*np.random.randn(1000)+0.4
res_a2 = 0.21*np.random.randn(1000)+0.3
res_b2 = 0.22*np.random.randn(1000)+0.3
```

```python
textstr1 = '\n'.join((
r'$\sigma_a=%.4f$' % (np.std(res_a1)),
r'$\sigma_b=%.4f$' % (np.std(res_b1))))

textstr2 = '\n'.join((
r'$\sigma_a=%.4f$' % (np.std(res_a2)),
r'$\sigma_b=%.4f$' % (np.std(res_b2))))
```

```python
#make a very nice looking plot 
fig, axes = plt.subplots(1,2,figsize =(10,3.5))
ax = axes [0]
ax.hist(res_a1, bins = 30, density = True, histtype = 'step')
ax.hist(res_b1, bins = 30, density = True, histtype = 'step')
ax.text(0.05,0.81, textstr1, transform=ax.transAxes, bbox = dict(facecolor ='white',edgecolor= 'black'))
ax.set_ylabel('Density')
fig.text(0.5,-0.04, '$\Delta E$',ha = 'center', fontsize = 20)
ax.set_title('Trial 1')

#now do the same thing for res_a2 and res_b2
ax = axes [1]
ax.hist(res_a2, bins = 30, density = True, histtype = 'step')
ax.hist(res_b2, bins = 30, density = True, histtype = 'step')
ax.text(0.05,0.81, textstr1, transform=ax.transAxes, bbox = dict(facecolor ='white',edgecolor= 'black'))
ax.set_title('Trial 2')
fig.text(0.5,-0.04, '$\Delta E$',ha = 'center', fontsize = 20)
plt.show()
```

```python
#generate some data 
_ = np.linspace (-1,1,100)
x,y = np.meshgrid(_,_)
z = x**2+x*y
```

```python
plt.contourf(x,y,z,levels= 30, vmax = 1.68)
plt.colorbar(label='asd')
#to adjust the color range, use vmin and vmax. can help w/ keeping color range in the arae that you're interested in.
```

```python
#plotting a pure contour plot (like you'd see on a map)
cs = plt.contour (x,y,z)
plt.clabel(cs, fontsize = 8)
```

```python
#3D surface plot, generally not preferred, however they tend to work well in animations 

fig, ax = plt.subplots(subplot_kw={'projection':'3d'})
ax.plot_surface(x,y,z,cmap ='coolwarm')
```

```python
 #STREAM PLOTS for plotting vector fields 
w =3 
_ = np.linspace(-3,3,100)
X,Y = np.meshgrid(_,_)
U = -1 -X**2 +Y
V = 1 + X - Y**2
speed = np.sqrt(U**2 + V**2)

#make a 4x4 plot to show all that you can do with this data

fig, axes = plt.subplots(2,2,figsize=(7,7))
ax = axes [0][0]
ax.streamplot(X,Y,U,V)
ax = axes [0][1]
ax.streamplot(X,Y,U,V, color =speed)
ax = axes [1][0]
lw = 5*speed / speed.max()
ax.streamplot(X,Y,U,V,linewidth = lw)
ax = axes [1][1]
lw =5*speed / speed.max()
seedpoints = np.array([[0,1],[1,0]])
ax.streamplot(X,Y,U,V, start_points=seedpoints)


    
```

```python
#ANIMATIONS 

#for example, below is a sin wave moving at 3 m/s, translate this into an animation

def f(x,t):
    return np.sin(x-3*t)
x = np.linspace(0,10*np.pi, 1000)
```

```python
plt.plot(x,f(x,0))
plt.plot(x,f(x,1.3))
```

```python
from matplotlib import animation 
from matplotlib.animation import PillowWriter 
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,4))
ln1, = plt.plot([],[])
time_text = ax.text(0.65,0.95,'', fontsize =15,  transform=ax.transAxes, bbox = dict(facecolor ='white',edgecolor= 'black'))
ax.set_xlim(0,10*np.pi)
ax.set_ylim (-1.5,-1.5)
def animate (i):
    ln1.set_data (x,f(x,1/50*i))
    time_text.set_text('t={:.2f}'.format(i/50))
ani = animation.FuncAnimation(fig,animate, frames = 240, interval = 50) 
ani.save('images/ani.gif', writer ='pillow', fps =50, dpi =100)
```

```python
#using animation function to animate 3D surface plot similar to the one that was plotted earlier 
_ = np.linspace(-1,1,100)
x,y = np.meshgrid(_,_)
z = x**2+x*y

plt.style.use(['default'])

#this data is pretty much copied and pasted from the above example on animation, but it's edited to fit the new example 
fig, ax = plt.subplots(subplot_kw ={'projection': '3d'})
ax.plot_surface(x,y,z,cmap = 'coolwarm', linewidth = 0,antialiased = False)
ax.view_init(elev = 10, azim = 0)


def animate (i):
    ax.view_init(elev = 10, azim = 3*i)
ani2 = animation.FuncAnimation(fig,animate, frames = 120, interval = 50) 
ani2.save('images/ani2.gif', writer ='pillow', fps =30, dpi =100)
```
